1. Program berikut adalah edge detection. Program ini berguna untuk mengidentifikasi kejadian kecelakaan. Dilihat dari contoh, bentuk truk berbeda jauh dari bentuk truk yang biasanya (normalnya). Teori visi komputer yang digunakan adalah edge detection. Algoritma edge Detection yang digunakan adalah Canny. Dalam algoritma Canny itu sendiri terdapat penggunaan filter Gaussian blur, menghitung gradient, pemilihan tepi, penerapan threshold (ambang batas objek), dan menganalisis tepi yang berhubungan.
<br><br>
Sebagai tambahan, dalam penyajian tampilan gambar, diatur agar menampilkan gambar awal dan gambar hasil dalam baris masing-masing.

```python
from matplotlib import pyplot as plt
import cv2

img_awal = cv2.imread("gambarUTS.jpg")
img = cv2.imread("gambarUTS.jpg", cv2.IMREAD_REDUCED_GRAYSCALE_4)
img = cv2.Canny(img, 200, 300)

fig = plt.figure(figsize = (100, 70))
fig.add_subplot(10, 1, 1)
plt.imshow(cv2.cvtColor(img_awal, cv2.COLOR_BGR2RGB))
fig.add_subplot(10, 1, 2)
plt.imshow(img)
```
